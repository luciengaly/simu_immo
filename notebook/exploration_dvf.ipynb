```python
# Import
import os
import glob
import pandas as pd
import plotly.express as px

pd.set_option("display.max_columns", 100)
```

```python
# Chemins
data_path = "../data/"
output_path = "../output/"
if not os.path.exists(output_path):
    os.makedirs(output_path)
```

```python
# Loading
if not os.path.exists(output_path + "toulouse.csv"):
    filepaths = glob.glob(data_path + "*.txt")
    dfs = []
    for filepath in filepaths:
        print(filepath)
        df = pd.read_csv(filepath, sep="|", header=0, low_memory=False)
        df["Valeur fonciere"] = df["Valeur fonciere"].str.replace(",", ".").astype(float)
        df["Surface Carrez du 1er lot"] = df["Surface Carrez du 1er lot"].str.replace(",", ".").astype(float)
        dfs.append(df.loc[df["Commune"] == "TOULOUSE"])
        
    df = pd.concat(dfs, axis=0, ignore_index=True)
    df.to_csv(output_path + "toulouse.csv", index=False, sep=";")
    
else:
    df = pd.read_csv(output_path + "toulouse.csv", sep=";", low_memory=False)
```

```python
# Macro filtres
df = df.loc[df["Type local"] == "Appartement"]
df = df.loc[df["Nature mutation"] == "Vente"]
df = df.loc[df["Nombre de lots"] == 1]
df = df.loc[~df["Surface Carrez du 1er lot"].isna()]
df["Prefixe de section"] = df["Prefixe de section"].astype(int)
df["Valeur fonciere"] = df["Valeur fonciere"].dropna()
df["Surface Carrez du 1er lot"] = df["Surface Carrez du 1er lot"].dropna()
df["VF/m2"] = df["Valeur fonciere"] / df["Surface Carrez du 1er lot"]
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# Supposons que df soit déjà défini et contienne les colonnes nécessaires


def subplot_boxplot(df, col, section, range_plan, subsection="AB"):
    # Création de la figure avec plusieurs sous-plots
    fig = make_subplots(
        rows=1,
        cols=6,
        subplot_titles=[
            "TLS - Tous les biens",
            "TLS - T1",
            "TLS - Downtown",
            "TLS - T1 - Downtown",
            f"TLS - T1 - Section {section} {subsection}",
            f"TLS - T1 - Plan {range_plan[0]}-{range_plan[1]}",
        ],
        shared_yaxes=True,
    )

    # TLS - Tous les biens
    df_tls = df.loc[(df["Code departement"] == 31) & (df["Code commune"] == 555)]
    fig.add_trace(go.Box(y=df_tls[col], name=""), row=1, col=1)

    # TLS - T1
    df_1p = df_tls.loc[df["Nombre pieces principales"] == 1]
    fig.add_trace(
        go.Box(y=df_1p[col], name=""),
        row=1,
        col=2,
    )

    # TLS - Downtown
    df_downtown = df_tls.loc[
        (
            df_tls["Prefixe de section"]
            .astype(str)
            .isin(["815", "816", "817", "818", "819"])
        )
        & (df_tls["Section"] == subsection)
    ]
    fig.add_trace(
        go.Box(y=df_downtown[col], name=""),
        row=1,
        col=3,
    )

    # TLS - T1 - Downtown
    df_downtown_1p = df_tls.loc[
        (
            df_tls["Prefixe de section"]
            .astype(str)
            .isin(["815", "816", "817", "818", "819"])
        )
        & (df_tls["Section"] == subsection)
        & (df_tls["Nombre pieces principales"] == 1)
    ]
    fig.add_trace(
        go.Box(y=df_downtown_1p[col], name=""),
        row=1,
        col=4,
    )

    # TLS - T1 - Section
    df_section = df_tls.loc[
        (df_tls["Prefixe de section"].astype(str) == section)
        & (df_tls["Section"] == subsection)
        & (df_tls["Nombre pieces principales"] == 1)
    ]
    fig.add_trace(
        go.Box(y=df_section[col], name=""),
        row=1,
        col=5,
    )

    # TLS - T1 - Plan
    df_plan = df_tls.loc[
        (df_tls["No plan"] >= range_plan[0])
        & (df_tls["No plan"] <= range_plan[1])
        & (df_tls["Section"] == subsection)
        & (df_tls["Nombre pieces principales"] == 1)
    ]
    fig.add_trace(
        go.Box(y=df_plan[col], name=""),
        row=1,
        col=6,
    )

    # Mise en forme du layout
    fig.update_layout(
        title=f"Distribution de la {col} avec différents filtres pour un appartement",
        showlegend=False,
        # height=700,
        # width=1800,
    )

    # Affichage
    fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
from typing import Tuple, Optional


class BoxPlotSubplot:
    def __init__(self, df: pd.DataFrame, col: str, section: str, range_plan: Tuple[int, int]):
        """
        Initialise le graphique avec les données et paramètres nécessaires.

        :param df: DataFrame contenant les données.
        :param col: Colonne de valeurs à afficher en boxplot.
        :param section: Code de section spécifique (ex: "818").
        :param range_plan: Intervalle des numéros de plan (ex: (100, 200)).
        """
        self.df = df
        self.col = col
        self.section = section
        self.range_plan = range_plan
        self.fig = make_subplots(
            rows=1,
            cols=6,
            subplot_titles=[
                "TLS - Tous les biens",
                "TLS - T1",
                "TLS - Downtown",
                "TLS - T1 - Downtown",
                f"TLS - T1 - Section {section} AB",
                f"TLS - T1 - Plan {range_plan[0]}-{range_plan[1]}",
            ],
            shared_yaxes=True,
        )

    def filter_data(self) -> dict:
        """Filtre les données en fonction des critères définis."""
        df_tls = self.df.loc[(self.df["Code departement"] == 31) & (self.df["Code commune"] == 555)]
        df_1p = df_tls.loc[df_tls["Nombre pieces principales"] == 1]

        downtown_sections = {"815", "816", "817", "818", "819"}
        df_downtown = df_tls.loc[df_tls["Prefixe de section"].astype(str).isin(downtown_sections) & (df_tls["Section"] == "AB")]
        df_downtown_1p = df_1p.loc[df_1p["Prefixe de section"].astype(str).isin(downtown_sections) & (df_1p["Section"] == "AB")]

        df_section = df_1p.loc[(df_1p["Prefixe de section"].astype(str) == self.section) & (df_1p["Section"] == "AB")]
        df_plan = df_1p.loc[
            (df_1p["No plan"].between(self.range_plan[0], self.range_plan[1]))
            & (df_1p["Section"] == "AB")
        ]

        return {
            "Tous les biens": df_tls,
            "T1": df_1p,
            "Downtown": df_downtown,
            "T1 - Downtown": df_downtown_1p,
            f"T1 - Section {self.section} AB": df_section,
            f"T1 - Plan {self.range_plan[0]}-{self.range_plan[1]}": df_plan,
        }

    def add_traces(self) -> None:
        """Ajoute les traces de boxplot aux subplots."""
        filtered_data = self.filter_data()

        for i, (title, data) in enumerate(filtered_data.items(), start=1):
            self.fig.add_trace(go.Box(y=data[self.col], name=""), row=1, col=i)

    def show(self) -> None:
        """Affiche le graphique."""
        self.fig.update_layout(
            title=f"Distribution de la {self.col} avec différents filtres",
            showlegend=False,
            height=700,
            width=1800,
        )
        self.fig.show()

    def save(self, filename: str, file_format: Optional[str] = "html") -> None:
        """
        Sauvegarde le graphe sous forme de fichier.

        :param filename: Nom du fichier sans extension.
        :param file_format: Format de sauvegarde ("html" ou "png").
        """
        if file_format == "html":
            self.fig.write_html(f"{filename}.html")
        elif file_format == "png":
            self.fig.write_image(f"{filename}.png")
        else:
            raise ValueError("Format non supporté. Choisissez 'html' ou 'png'.")
```

```python
# ====== UTILISATION ======

# Initialisation de l'objet
plotter = BoxPlotSubplot(df, col="Valeur fonciere", section="818", range_plan=(100, 200))

# Ajout des données et affichage
plotter.add_traces()

# Sauvegarde du graphe
plotter.save("boxplot_graph_VF", file_format="html")  # Sauvegarde en HTML



# Initialisation de l'objet
plotter = BoxPlotSubplot(df, col="VF/m2", section="818", range_plan=(400, 460))

# Ajout des données et affichage
plotter.add_traces()

# Sauvegarde du graphe
plotter.save("boxplot_graph_VFm2", file_format="html")  # Sauvegarde en HTML
```
